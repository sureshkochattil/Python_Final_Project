```python
import pandas as pd
import numpy as np
import statistics
import matplotlib.pyplot as plt
```

```python
breast_cancer_file = pd.read_csv('Breast-Cancer-Wisconsin.csv')
```

```python
breast_cancer_file_replace = breast_cancer_file.replace({'A7':'?'},np.NaN)
```

```python
converted_numeric = pd.to_numeric(breast_cancer_file_replace['A7'])
```

```python
is_NaN = converted_numeric.isnull()
```

```python
print(is_NaN.sum())
```

```python
converted_mean = converted_numeric.fillna(converted_numeric.mean(skipna = True))
```

```python
statistics.mean(converted_mean)
```

```python
breast_cancer_file_replace_all = breast_cancer_file.replace('?',np.NaN)
```

```python
convert_numeric_all = breast_cancer_file_replace_all.apply(pd.to_numeric)
```

```python
convert_mean_all = convert_numeric_all.fillna(convert_numeric_all.mean(skipna = True))
```
